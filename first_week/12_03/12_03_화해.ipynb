```python
def get_slug(pid):
    import re, requests

    url = f"http://www.hwahae.co.kr/goods/{pid}"
    html = requests.get(url).text  # ← 변수명 통일

    m = re.search(r'"asPath":"\\/goods\\/(.*?)"', html)
    if m:
        return m.group(1)

    print("❌ slug 못 찾음:", pid)
    return None
```

```python
def fetch_json(driver, pid, token): 
    slug = get_slug(pid)
    if slug is None:
        print("❌ slug 없음:", pid)
        return None
    url = (
        f"https://www.hwahae.co.kr/_next/data/"
        f"{token}/goods/{slug}/{pid}.json"
        f"?id={pid}&goods_tab=review_ingredients"
    )

    print("요청 URL:", url)  # 디버그용

    driver.get(url)
    time.sleep(1)

    try:
        pre = driver.find_element("tag name", "pre").text
        return json.loads(pre)
    except Exception as e:
        print("JSON 파싱 실패:", pid, e)
        return None
```

### 관련 제품 ID 자동 수집 

```python
def extract_related_ids(data):
    ids = set()

    def scan(obj):
        if isinstance(obj, dict):
            for v in obj.values():
                scan(v)

        elif isinstance(obj, list):
            for v in obj:
                scan(v)

        elif isinstance(obj, int):
            # 화해 상품 ID 범위(대략 5~6자리 숫자)
            if 10000 < obj < 9999999:
                ids.add(obj)

    scan(data)
    return list(ids)
```

### 전체 크롤링 구조 

```python
def crawl_related(start_id, limit=50): 
    token = get_next_token()
    driver = get_driver()

    visited = set()
    queue = [start_id]
    saved = []

    while queue and len(saved) < limit: 
        pid = queue.pop(0)
        if pid in visited: 
            continue
        visited.add(pid)
        print("수집 중", pid)

        data = fetch_json(driver, pid, token)
        if data is None: 
            print("요청 실패", pid)
            continue
        saved.append((pid, data))

        rel = extract_related_ids(data)
        for nid in rel: 
            if nid not in visited: 
                queue.append(nid)
        time.sleep(0.3)

    driver.quit()
    return saved
```

#### Selenium 기반 완전 자동 수집 

```python
import json
import os
import time
import re
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()  

def fetch_json_by_pid(pid: int):
    """goods/{pid} 페이지에서 __NEXT_DATA__ JSON만 뽑기"""
    url = f"https://www.hwahae.co.kr/goods/{pid}"
    print("열기:", url)
    driver.get(url)
    time.sleep(1.0)  # 로딩 대기 

    try:
        script = driver.find_element(By.ID, "__NEXT_DATA__").get_attribute("textContent")
        data = json.loads(script)
        return data
    except Exception as e:
        print("JSON 추출 실패:", pid, e)
        return None

def extract_related_ids(data):
    """JSON 전체를 훑어서 '그럴듯한 숫자 id' 를 수집 (네가 만들던 C 단계)"""
    ids = set()

    def scan(obj):
        if isinstance(obj, dict):
            for v in obj.values():
                scan(v)
        elif isinstance(obj, list):
            for v in obj:
                scan(v)
        elif isinstance(obj, int):
            # 대충 상품 id / 관련 id 로 보이는 숫자 범위만 필터
            if 10000 < obj < 99999999:
                ids.add(obj)

    scan(data)
    return list(ids)

def crawl_related(start_id, limit=30):
    """BFS 로 start_id 기준으로 JSON 계속 수집"""
    visited = set()
    queue = [start_id]
    saved = []

    while queue and len(saved) < limit:
        pid = queue.pop(0)
        if pid in visited:
            continue
        visited.add(pid)
        print("수집 중", pid)

        data = fetch_json_by_pid(pid)
        if data is None:
            print("요청 실패", pid)
            continue

        # (pid, data) 같이 저장해 둬야 함
        saved.append((pid, data))

        
        rel_ids = extract_related_ids(data)
        for nid in rel_ids:
            if nid not in visited:
                queue.append(nid)

        time.sleep(0.5)  

    return saved

def save_json_list(data_list, folder="product"):
    """(pid, data) 리스트를 product/{pid}.json 으로 저장"""
    os.makedirs(folder, exist_ok=True)
    for pid, data in data_list:
        path = os.path.join(folder, f"{pid}.json")
        with open(path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=2)
    print("저장 완료", len(data_list))


start_id = 73532  
data_list = crawl_related(start_id, limit=30)
save_json_list(data_list)


driver.quit()
```

```python
import json
import pandas as pd
import glob
import os

# 1) 한 개 JSON에서 기능 비율 뽑는 함수 (이미 썼던 로직 정리 버전)
def extract_purpose_ratio_from_data(data):
    product = data["pageProps"]["productGoodsPairData"]["product"]
    ingredients = data["pageProps"]["productIngredientInfoData"]["ingredients"]

    brand = product["brand"]["name"]
    name = product["name"]

    rows = []
    for ing in ingredients:
        purposes = ing.get("purposes", [])
        if purposes:
            for p in purposes:
                rows.append(p["group_name"])
        else:
            rows.append("기능 없음")

    s = pd.Series(rows).value_counts(normalize=True).mul(100).round(1)
    # dict로 바꿔서 메타 붙이기
    d = s.to_dict()
    d["brand"] = brand
    d["product"] = name
    return d

# 2) 폴더에 있는 여러 JSON 대상 일괄 적용
records = []
for path in glob.glob("product/*.json"):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    records.append(extract_purpose_ratio_from_data(data))

df = pd.DataFrame(records)

# 3) 아모레 계열 flag 붙이기
amore = ["아모레퍼시픽", "라네즈", "마몽드", "이니스프리", "헤라", "프리메라"]
df["is_amore"] = df["brand"].apply(lambda x: x in amore)

print(df.head())
```
