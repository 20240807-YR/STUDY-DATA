```python
from pytrends.request import TrendReq
import pandas as pd
import time

# 재시도 + 백오프 설정
pytrend = TrendReq(
    hl='ko',
    tz=540,
    retries=5,         # 최대 5번 재시도
    backoff_factor=2   # 2, 4, 8, 16, 32초 이런 식으로 대기 늘어남
)

keywords = ["skincare", "sunscreen", "cleansing", "makeup"]
```

```python
def safe_interest_over_time(kw_list, timeframe='2024-01-01 2025-01-01', geo='KR'):
    while True:
        try:
            pytrend.build_payload(kw_list=kw_list, timeframe=timeframe, geo=geo)
            df = pytrend.interest_over_time()
            return df
        except Exception as e:
            print("에러 발생:", e)
            print("조금 쉬었다가 다시 시도...")
            time.sleep(60)   # 60초 정도 대기 후 재시도
```

```python
from pytrends.request import TrendReq
import time

pytrend = TrendReq(hl='ko', tz=540)

keywords = ["skincare", "sunscreen", "cleansing", "makeup"]

result = {}

for kw in keywords:
    time.sleep(5)  # 반드시 쉬어야 함
    pytrend.build_payload([kw], geo='KR', timeframe='today 12-m')
    df = pytrend.interest_over_time()
    result[kw] = df
    print(f"{kw} 완료")
```

```python
df = safe_interest_over_time(
    kw_list=keywords,
    timeframe='2024-01-01 2025-01-01',
    geo='KR'
)

print(df.head())
```
